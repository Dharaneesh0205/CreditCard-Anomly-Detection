```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler


data = pd.read_csv('C:/Users/dhara/Downloads/creditcard.csv/creditcard.csv')

# View basic information
print(data.info())
print(data.describe())

# Check for missing values
if data.isnull().sum().any():
    data = data.dropna()

# Feature scaling
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.drop(['Class'], axis=1))
data_scaled = pd.DataFrame(data_scaled, columns=data.columns[:-1])
data_scaled['Class'] = data['Class']
```

```python
from scipy.stats import zscore

# Calculate Z-scores for each feature
z_scores = np.abs(zscore(data_scaled.drop(['Class'], axis=1)))
threshold = 3  # Common threshold for Z-score

# Detect anomalies
z_score_anomalies = (z_scores > threshold).any(axis=1)
print("Z-score based anomalies detected:", z_score_anomalies.sum())
```

```python
from scipy.spatial.distance import mahalanobis
from numpy.linalg import inv

def mahalanobis_distance(data, mean, inv_cov):
    diff = data - mean
    return np.sqrt(diff @ inv_cov @ diff.T)

# Calculate the covariance matrix and its inverse
mean = data_scaled.drop(['Class'], axis=1).mean().values
cov_matrix = np.cov(data_scaled.drop(['Class'], axis=1), rowvar=False)
inv_cov_matrix = inv(cov_matrix)

# Calculate Mahalanobis distance
mahalanobis_distances = data_scaled.drop(['Class'], axis=1).apply(
    lambda x: mahalanobis(x, mean, inv_cov_matrix), axis=1)
threshold_md = mahalanobis_distances.quantile(0.99)

# Detect anomalies
md_anomalies = mahalanobis_distances > threshold_md
print("Mahalanobis Distance based anomalies detected:", md_anomalies.sum())
```

```python
from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(n_neighbors=20, contamination=0.01)
lof_labels = lof.fit_predict(data_scaled.drop(['Class'], axis=1))
lof_anomalies = lof_labels == -1
print("LOF anomalies detected:", sum(lof_anomalies))
```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(contamination=0.01, random_state=42)
iso_forest.fit(data_scaled.drop(['Class'], axis=1))
iso_labels = iso_forest.predict(data_scaled.drop(['Class'], axis=1))
iso_anomalies = iso_labels == -1
print("Isolation Forest anomalies detected:", sum(iso_anomalies))
```

```python
from sklearn.svm import OneClassSVM

one_class_svm = OneClassSVM(nu=0.01, kernel='rbf', gamma='scale')
svm_labels = one_class_svm.fit_predict(data_scaled.drop(['Class'], axis=1))
svm_anomalies = svm_labels == -1
print("One-Class SVM anomalies detected:", sum(svm_anomalies))
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming 'Class' = 1 indicates an anomaly
y_true = data_scaled['Class'].values

# List to store results
results = []

# Helper function for evaluation
def evaluate_model(anomalies_detected, method_name):
    y_pred = anomalies_detected.astype(int)
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred) 
    f1 = f1_score(y_true, y_pred)
    results.append([method_name, accuracy, precision, recall, f1])

# Evaluate each model
evaluate_model(z_score_anomalies, "Z-score")
evaluate_model(md_anomalies, "Mahalanobis Distance")
evaluate_model(lof_anomalies, "LOF")
evaluate_model(iso_anomalies, "Isolation Forest")
evaluate_model(svm_anomalies, "One-Class SVM")

# Convert results to a DataFrame
results_df = pd.DataFrame(results, columns=["Method", "Accuracy", "Precision", "Recall", "F1 Score"])

# Display results as a table
print("Anomaly Detection Results:")
print(results_df)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot anomalies identified by each method
fig, axes = plt.subplots(3, 2, figsize=(15, 12))
methods = [z_score_anomalies, md_anomalies, lof_anomalies, iso_anomalies, svm_anomalies]
titles = ["Z-score", "Mahalanobis", "LOF", "Isolation Forest", "One-Class SVM"]

for ax, anomalies, title in zip(axes.flat, methods, titles):
    sns.scatterplot(data=data_scaled, x='V1', y='V2', hue=anomalies, palette=['blue', 'red'], ax=ax)
    ax.set_title(title)

plt.tight_layout()
plt.show()
```
